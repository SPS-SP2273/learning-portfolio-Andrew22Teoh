## 1 A simple plot

### 1.1 Letâ€™s look at some code.

```python
from matplotlib import pyplot as plt

# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

# Let's start plotting
plt.plot(x, y_1, color='red', linestyle='dashed', label='Y values')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha = 1)
plt.legend(loc='center')
plt.show()
```

### 1.2 Adding another plot

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

# Let's start plotting
plt.plot(x, y_1, color='red', linestyle='dashed', label='Y_1 values')
plt.plot(x, y_2, c = "green", ls = "dotted", lw = 2, label = "Y_2 values")
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y_1 and Y_2')
plt.grid(alpha = 1)
plt.legend(loc='lower center')
plt.show()
```

### 1.3 Yet another plot but with error bars

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6, 7]
y_1 = [0, 2, 4, 6, 8, 10, 12, 14]
y_2 = [0, 3, 6, 9, 12, 15, 18, 21]
y_3 = [0, 4, 8, 12, 16, 20, 24, 50]
x_error, y_error = .1, 0.75

# Let's start plotting
plt.plot(x, y_1, color='red', linestyle='dashed', label='Y_1 values')
plt.plot(x, y_2, c = "green", ls = "dotted", lw = 2, label = "Y_2 values")
plt.errorbar(x, y_3, xerr = x_error, yerr = y_error, c = "blue", label = "Y_3 with error bars")
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y_1, Y_2, and Y_3')
plt.grid(alpha = 1)
plt.legend(loc='lower center')
plt.show()
```

### 1.4 Experiment, Experiment, Experiment

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6, 7]
y_1 = [0, 2, 4, 6, 8, 10, 12, 14]
y_2 = [0, 3, 6, 9, 12, 15, 18, 21]
y_3 = [0, 4, 8, 12, 16, 20, 24, 50]
x_error, y_error = .1, 5

# Let's start plotting
plt.plot(x, y_1, color='#fa9fb5', linestyle='dashed', label='Y_1 values')
plt.plot(x, y_2, c = "#c51b8a", ls = "dotted", lw = 2, label = "Y_2 values")
plt.errorbar(x, y_3, xerr = x_error, yerr = y_error, c = "#feb24c", label = "Y_3 with error bars")
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title(label = "X against Y_1, Y_2, Y_3", c = "#f03b20")
plt.grid(alpha = 1, c = "#756bb1", ls = "dotted")
plt.legend(loc = "bet")
plt.show()
```

1. New colours yay!!
2. alpha specifies the opacity of the gridlines in the background, ranging from \[0,1\] with 0 as no gridlines and 1 as very opaque gridlines.
3. The colour of the title can be changed with {c = "{colour}"} after the label is defined.
4. Very Interesting grid has been made
5. The legend location has been changed! There are multiple options, including the obvious upper/center/lower, left/right/center, as well as "best" for the program to automatically do it for you.

### 1.5 Better with NumPy

```python
import numpy as np
x = [0,1,2,3,4,5,6]
y_1 = [0,2,4,6,8,10,12]
y_2 = [0,3,6,9,12,15,18]

np_x = np.array(x)
np_y_1 = np.array(y_1)
np_y_2 = np.array(y_2)

plt.plot(np_x, np_y_1, c = "red", ls = "dashed", lw = 50, label = "Y_1")
plt.plot(np_x, np_y_2, c = "green", ls = "dashed", lw = .2, label = "Y_2")
plt.grid(alpha = .5)
plt.show()
```

### 1.6 Adding mathematical functions

```python
x = np.array([0,1,2,3,4,5,6])
x_2 = x**2
sin_x = np.sin(x)

plt.plot(x, x_2, c = "blue", label = "x_2", ls = "dotted")
plt.plot(x, sin_x, c = "purple", label = "sin_x", ls = "dotted")
plt.grid(alpha = 0.3)
```

### 1.7 We need another axis!

```python
x = np.linspace(0,6,100)
#Function is called linespace, where it gives you 100 evenly spaced data points between 0 and 6

x_2 = x**2
sin_x = np.sin(x)
plt.plot(x, x_2, c = "pink", ls = "dotted", label = "x**2")
plt.grid(alpha = 0.2)
plt.legend(loc = "upper center")

plt.twinx()
plt.plot(x, sin_x, color = "purple", ls= "dotted", label = "sin(x)")

plt.legend(loc = "lower center")
plt.show()
```

## 2 A real example: Global Warming

### 2.1 Plotting data from files

```python
data = np.loadtxt("Complete_TAVG_daily.txt", skiprows = 24)
#recall that slicing 2D numpy arrays can be done as np_array[a, b] instead of np_array[a][b]
date = data[:, 0]
anomaly = data[:, -1]
```

```python
plt.plot(date, anomaly, alpha = .4, c = "purple", label = "Temperature Anomalies")
plt.ylim([np.min(anomaly)-.5,np.max(anomaly)+.5])
plt.hlines(0, date[0], date[-1], ls = "dashed", color = "grey")
plt.ylabel("Temperatuer Anomaly")
plt.xlabel("Year")
plt.legend()
plt.title("Temperature Anomalies over the years")
plt.show()
```

```python
plt.style.use('fivethirtyeight')
plt.plot(date, anomaly, alpha = .4, c = "purple", label = "Temperature Anomalies")
plt.ylim([np.min(anomaly)-.5,np.max(anomaly)+.5])
plt.hlines(0, date[0], date[-1], ls = "dashed", color = "grey")
plt.ylabel("Temperatuer Anomaly")
plt.xlabel("Year")
plt.legend()
plt.title("Temperature Anomalies over the years")
plt.show()
```

### 2.2 xkcd!

```python
plt.style.use('default')
plt.xkcd()
plt.plot(date, anomaly, alpha = .4, c = "purple", label = "Temperature Anomalies")
plt.ylim([np.min(anomaly)-.5,np.max(anomaly)+.5])
plt.hlines(0, date[0], date[-1], ls = "dashed", color = "grey")
plt.ylabel("Temperatuer Anomaly")
plt.xlabel("Year")
plt.legend()
plt.title("Temperature Anomalies over the years")
plt.show()
plt.style.use('default')
```

## Exercise 2 :  A simple mathematical plot

```python
x = np.linspace(0,50,10000)
y = np.sin(x)*(np.e)**(-x/10)

plt.plot(x,y, c = "darkblue")
plt.ylim([-1,1])
plt.grid(alpha = 0.25)
plt.xlabel("x", fontsize = 15)
plt.ylabel("y", fontsize = 15, rotation = 0)
plt.title(r'$y = e^{-x/10}sin(x)$')



plt.show()
```
