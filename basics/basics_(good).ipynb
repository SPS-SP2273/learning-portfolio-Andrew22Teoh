## 1 More Python Basics

### 1.1 Computers read = from Right to Left!

```python
a = 10
b = a * 2 + 5 #this sets b = 25
b = b * 4 #this sets b = 100
print (b)
```

### 1.2 Shorter and cleaner

```python
a = 10
a += 10 #sets a = 20
a *= 2 # sets a = 40
a -= 30 #sets a = 10
a /= 2 #sets a = 5
a //= 2 #sets a = 2
a **= 4 #sets a = 16
a %= 3 #sets a = 1
print (a)
```

### 1.3 Asking questions

```python
superheroes = ["batman", "martian", "silver surfer", "spiderman", "wonder woman", "black widow"]
if "black widow" in superheroes:
    print ("I love empowered female characters!")
elif "homelander" in superheroes:
    print ("Superheroes are an antiquated artifact of a war culture that should no longer exist")
else:
    print ("eh idk i guess comic books are whatever")
```

### 1.4 There is more to if

```python
a = 501
if a%5==0:
    print ("a is a multiple of 5")
elif a%12 == 0:
    print ("a is a multiple of 12")
elif a%4==0:
    print("a is a multiple of 4")
elif a%2==0:
    print ("a is an even number")
else:
    print ("a is a weird sad odd number")
```

### 1.5 Python stores information in different formats or types.

```python
x = 1.583
print (type(x)) #printing a float x
x = str(x) #typecasting x into a string
print(type(x)) #printing a str x
x = complex(x) #typecasting x into a complex number
print(type(x)) #printing a complex number x
x = int(float(x)) #typecasting x from a complex number into a float is not possible
print(type(x)) #printing an integer x
```

```python
greeting = "Hello World!"
print(f"{greeting:<1000}")
print(f'{greeting:^100}')          #makes a text block 100 characters long, and puts the greeting in the middle of this block
print(f'{greeting:^100000}')       #block larger than available space, fails to center anything
print(f'{greeting:^30}')           #makes a block of text 30 characters long, and centers greeting in the middle of that
print(f'{greeting:>30}')
print(f'{4*"123456789012345678901234567890"}')
```
